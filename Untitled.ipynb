```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the dataset
excel_file_path = 'Dry_Bean_Dataset.xlsx'
data = pd.read_excel(excel_file_path)

# Separating features (x) and target variable (y)
x = data.drop('Class', axis=1)
y = data['Class']

# Encoding the categorical target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Splitting the data into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Normalize the features
scaler = StandardScaler()
train_x_normalized = scaler.fit_transform(train_x)
test_x_normalized = scaler.transform(test_x)

categoriesArr = ["Attempt", "Hidden Layer Size", "Solver", "Learning_rate_init", "Max_iter", "Training Accuracy", "Testing Accuracy"]
# Table to store results
resultsArr = []

# Attempt 1
#define the artificial neural network
ann = MLPClassifier(hidden_layer_sizes=(4,), solver='sgd', learning_rate_init=0.8, max_iter=200, random_state = 42)
#train the dataset
ann.fit(train_x_normalized, train_y)
# get train and test accuracy thru accuracy score function
train_accuracy = accuracy_score(train_y, ann.predict(train_x_normalized))
test_accuracy = accuracy_score(test_y, ann.predict(test_x_normalized))
resultsArr.append([1, 4, 'Sgd', 0.8, 200, train_accuracy, test_accuracy])

plt.plot(range(1, ann.n_iter_ + 1), ann.loss_curve_, label='Training')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Training Loss Curve (Attempt 1)')
plt.legend()
plt.show()

for i in range(len(categoriesArr)):
    for j in range(len(resultsArr)):
        print(categoriesArr[i], ":", resultsArr[j][i])
```
